```python
import pandas as pd
import numpy as np
Chocolate_Store={"store" : [1,2,3,4,5,6] , "price":[20,16,15,16,13,10] , "unit_sold":[0,3,7,4,6,10]}
formate=pd.DataFrame(Chocolate_Store)
print(formate)
```

#### Hint: to make your life easier, you can use `seaborn' library for the scatter plots. there is a function called 'pairplot' that can help you with this task.

```python
import matplotlib.pyplot as mp
mp.scatter(formate["price"]  ,formate [ "unit_sold"])
mp.xlabel('price')
mp.ylabel('unit_sold')
mp.show()
```

```python
price=np.array(formate["price"])
unit_sold=np.array(formate["unit_sold"])


x_mean= np.mean(price)
y_mean=np.mean(unit_sold)
print (x_mean , y_mean)
#print(price)
def calculation_of_b1(x,y):
 upper=np.sum((x-x_mean)*(y-y_mean))
 lower=np.sum((x-x_mean)**2)
 print(upper ,lower)
 print (upper/lower)
 return upper/lower

b1=calculation_of_b1(price, unit_sold)
b0=y_mean-(b1)*x_mean
print(b0)
regression_line=b1+(b0*price)
print(regression_line)
```

```python
mp.scatter(price , unit_sold , s=20 , c='blue' , alpha=1 ,label='Data Points' )
mp.plot(price , regression_line , c='red',label='Regression Line')
mp.legend()
mp.xlabel('price')
mp.ylabel('unit_sold')
mp.show()
```
